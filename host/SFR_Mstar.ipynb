```python
def load_Schulze21():
    colnames = "Redshift        chi2  dof   "+\
                "Ebv_host        Ebv_host+  Ebv_host-  M_FUV   M_FUV+  M_FUV-  "+\
                "M_B     M_B+    M_B-    M_Ks    M_Ks+   M_Ks-   "+\
                "lgSFR   lgSFR+  lgSFR-  lgM     lgM+    lgM-    "+\
                "lgsSFR  lgsSFR+ lgsSFR- Age Age+ Age-"
    colnames = colnames.split(" ")
    while "" in colnames:
            colnames.remove("")
    
    f = open(workdir+"data/catalogs/Schulze2021/table4.dat")
    lines = f.readlines()
    f.close()
    
    nline = len(lines)
    dt = np.zeros((nline, 27))
    
    tb = Table(dt)
    
    names = []
    types = []
    for i in range(nline):
        line = lines[i].split(" ")
        while "" in line:
            line.remove("")
        while "\n" in line:
            line.remove("\n")
        if len(line) == 30:
            newline = [line[0], line[1]+" "+line[2]]+line[3:]
        else:
            newline = line
            
        for j in range(len(newline)):
            if j not in [0,1]:
                tb["col%d"%(j-2)][i] = newline[j]
        
            elif j==0:
                names.append(newline[j])
            elif j==1:
                types.append(newline[j])
    
    for i in range(27):
        tb.rename_column("col%d"%i, colnames[i])
    
    tb["ID"] = names
    tb["Type"] = types
    return tb


def add_ccsn(ax):
    df = load_Schulze21()
    print (len(df), "CCSNe from iPTF/PTF")
    
    subix = df["lgM"] < np.log10(2e+9)
    subdf = df[subix]
    subdt = subdf["lgSFR"].data - subdf["lgM"].data
    ind_low = subdt < np.log10(8e-11)
    #ind_low = subdt < np.log10(1e-10)
    np.sum(ind_low) / len(ind_low)
    
    ngridx = 40
    ngridy = 50
    x = np.linspace(xmin, xmax, ngridx+1)
    y = np.linspace(ymin, ymax, ngridy+1)
    X, Y = np.meshgrid(x, y, sparse=True)
    Z = np.zeros((ngridy+1, ngridx+1))
    
    for i in range(len(df)):
        mu1 = df["lgM"][i]
        sigma1 = df["lgM+"][i]/2 +  df["lgM-"][i]/2
        mu2 = df["lgSFR"][i]
        sigma2 = df["lgSFR+"][i]/2 +  df["lgSFR-"][i]/2
        #print (mu1, sigma1, mu2, sigma2)
        z = 1 / (sigma1 * np.sqrt(2 * np.pi)) * np.exp(-1 * (X - mu1)**2 / (2 * sigma1**2)) *\
            1 / (sigma2 * np.sqrt(2 * np.pi)) * np.exp(-1 * (Y - mu2)**2 / (2 * sigma2**2))
        nbad = np.sum(np.isnan(z))
        if nbad != 0:
            print (i, nbad)
            z[np.isnan(z)] = 0
        else:
            Z += z
    
    cmap = "Greys"
    cmap = plt.cm.binary
    maxz = max(Z.ravel())
    minz = min(Z.ravel())
    c1 = 195 # 38%
    c2 = 74 # 68%
    c3 = 23 # 87%
    c4 = 8 # 95%
    levels =  [c4, c3, c2, c1, maxz] 
    vtotal = np.sum(Z)
    
    v1 = np.sum(Z[Z>c1])
    v2 = np.sum(Z[Z>c2])
    v3 = np.sum(Z[Z>c3])
    v4 = np.sum(Z[Z>c4])
    print ("%.2f, %.2f, %.2f, %.2f"%(v1/vtotal, v2/vtotal, v3/vtotal, v4/vtotal))
    ax.contour(x, y, Z, alpha = 1, zorder = 2,
                cmap = cmap, levels=levels,
                vmin=-50, vmax=maxz)
```

```python
def add_SFR_MS_R15(ax, xs, color = "k"):
    c1 = np.random.normal(0.76, 0.01, 1000) 
    c3 = np.random.normal(7.64, 0.02, 1000) 
    myas = c1
    mybs =  -1*c3
    As = myas.reshape(len(myas), 1)
    Bs = mybs.reshape(len(mybs), 1)
    Xs = xs.reshape(1, len(xs))
    ysv = np.matmul(As, Xs) + Bs #- Xs
    ysmax = np.percentile(ysv, [95], axis = 0)[0]
    ysmin = np.percentile(ysv, [5], axis = 0)[0]
    ax.fill_between(xs, ysmax, ysmin, zorder = 1, color = color, alpha = 0.5)
    
    
def custom_ax(ax):
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    ax.set_xlabel("log"+r"$_{10}(M_\ast/M_\odot)$")
    ax.set_ylabel("log"+r"$_{10}{\rm SFR}$"+" ("+r"$M_\odot\,{\rm yr^{-1}})$")
    ax.tick_params(which = 'major', length = 4, top=True, direction = "in", right = True)
    ax.tick_params(which = 'minor', length = 2, top=True, direction = "in", right = True)
    ax.xaxis.set_major_locator(plt.MultipleLocator(1))
    ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))
    ax.yaxis.set_major_locator(plt.MultipleLocator(1))
    ax.yaxis.set_minor_locator(plt.MultipleLocator(0.1))
    
    # add constant sSFR
    color = "steelblue"
    sSFRs = np.array([1e-8, 1e-9, 1e-10, 1e-11])
    xs = np.linspace(xmin, xmax)
    for i in range(len(sSFRs)):
        sSFR = sSFRs[i]
        ys = xs + np.log10(sSFR)
        ax.plot(xs, ys, ":", color = color, zorder = 3,linewidth = 0.6)
        ind = int(np.log10(sSFR))
        mytext = r"$10^{%d}\,{\rm yr^{-1}}}$"%(ind)
        xpos = 9
        if ind==-8:
            xpos = 9
            mytext = r"${\rm sSFR = 10^{%d}\,{\rm yr^{-1}}}$"%(ind)
        elif ind==-9:
            xpos = 10.3
        elif ind==-10:
            xpos = 10.7
        elif ind==-11:
            xpos = 10.7
        ypos = xpos + np.log10(sSFR)+0.2
        ax.text(xpos, ypos, mytext, zorder = 3,
                color = color, fontsize = fs-1, rotation = 48)
        
    # add main sequence star-forming galaxy
    # SDSS, Z=0, Eq(5) of Elbas+2007
    ys1 = np.log10(8.7 * (10**xs/1e+11)**0.77) 
    ys2 = np.log10((8.7+7.4) * (10**xs/1e+11)**0.77) 
    ys3 = np.log10((8.7-3.7) * (10**xs/1e+11)**0.77) 
    ysv = np.vstack([ys1, ys2, ys3])
    ysmax = np.max(ysv, axis = 0)
    ysmin = np.min(ysv, axis = 0)
    #ax.fill_between(xs, ysmax, ysmin, zorder = 1, color = "yellow", alpha = 0.5)
    #ax.plot(xs, ys1, color = "yellow", linestyle = "--")
    
    # Peng+2015, z~0
    ys = 0.76*xs-7.64
    add_SFR_MS_R15(ax, xs, color = "lightgreen")
    
    # GOODS, Z=1, Eq(4) of Elbas+2007
    ys1 = np.log10(7.2 * (10**xs/1e+10)**0.9) 
    ys2 = np.log10((7.2+7.2) * (10**xs/1e+10)**0.9) 
    ys3 = np.log10((7.2-3.6) * (10**xs/1e+10)**0.9) 
    ysv = np.vstack([ys1, ys2, ys3])
    ysmax = np.max(ysv, axis = 0)
    ysmin = np.min(ysv, axis = 0)
    ax.fill_between(xs, ysmax, ysmin, zorder = 1, color = "yellow",alpha = 0.5)
```

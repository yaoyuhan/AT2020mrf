```python
import numpy as np
import pandas as pd
import glob
from astropy.table import Table
from astropy.wcs import WCS
from astropy import units as u
from astropy.io import fits
from astropy.coordinates import SkyCoord
from astropy.visualization import make_lupton_rgb

import matplotlib
import matplotlib.pyplot as plt
fs = 10
matplotlib.rcParams['font.size']=fs

from astropy.cosmology import FlatLambdaCDM
cosmo = FlatLambdaCDM(H0=70., Om0=0.3)
```

```python
lambda_effs = np.array([4754, 6175, 7711, 8898, 9762], dtype = float)
filts = ["g", "r", "i", "z", "y"]


def add_srg_circle(ax, mywcs):

    ra_srg = 236.975833
    dec_srg = 44.651944
    r98_srg = 4.28/3600 * u.deg
    coo_srg = SkyCoord(ra=ra_srg*u.degree, dec=dec_srg*u.degree, frame='icrs')
    position_angles = np.linspace(0, 360, 360) * u.deg
    coo_srg_circles = coo_srg.directional_offset_by(position_angles, r98_srg)  
    xcircles_srg, ycircles_srg = mywcs.wcs_world2pix(coo_srg_circles.ra.degree, 
                                             coo_srg_circles.dec.degree, 0)
    ax.plot(xcircles_srg, ycircles_srg, linestyle=':', color = "white", linewidth=1.)
    ax.text(xcircles_srg[-45], ycircles_srg[-45], "$SRG$"+" position", color = "white")
    
    
def add_cxo_circle(ax, mywcs):
    # obsID 25050
    ra_cxo = 236.9757561201018
    dec_cxo = 44.65217469439337
    r98_cxo = 1.47/3600 * u.deg
    coo_cxo = SkyCoord(ra=ra_cxo*u.degree, dec=dec_cxo*u.degree, frame='icrs')
    position_angles = np.linspace(0, 360, 360) * u.deg
    coo_cxo_circles = coo_cxo.directional_offset_by(position_angles, r98_cxo)  
    xcircles_cxo, ycircles_cxo = mywcs.wcs_world2pix(coo_cxo_circles.ra.degree, 
                                             coo_cxo_circles.dec.degree, 0)
    ax.plot(xcircles_cxo, ycircles_cxo, linestyle='-.', color = "gold", linewidth=0.9)
    ax.text(xcircles_cxo[-60], ycircles_cxo[-60], "$Chandra$"+" positions", color = "gold")
    
    # obsID 25064
    ra_cxo = 236.97575996086582
    dec_cxo = 44.65198976874989
    r98_cxo = 0.82/3600 * u.deg
    coo_cxo = SkyCoord(ra=ra_cxo*u.degree, dec=dec_cxo*u.degree, frame='icrs')
    position_angles = np.linspace(0, 360, 360) * u.deg
    coo_cxo_circles = coo_cxo.directional_offset_by(position_angles, r98_cxo)  
    xcircles_cxo, ycircles_cxo = mywcs.wcs_world2pix(coo_cxo_circles.ra.degree, 
                                             coo_cxo_circles.dec.degree, 0)
    ax.plot(xcircles_cxo, ycircles_cxo, linestyle='--', color = "gold", linewidth=1.)
```

```python
    z = 0.1353
    
    dA_Mpc = cosmo.angular_diameter_distance(z = z).value # in Mpc
    d_kpc = dA_Mpc * 0.5 / (60*60) / 180 * np.pi * 1e+3 # in kpc
    print ("%.1f kpc"%d_kpc)
    
    # ZTF location
    ra =  236.9756899
    dec = 44.6520401
    hw = 150
    
    files = glob.glob("../data/data_20mrf/host/hsc/*.fits")
    files = np.array(files)
    files = files[np.argsort(files)]
    
    fileg = files[0]
    filei = files[1]
    filez = files[4]
    
    hdug = fits.open(fileg)[1]
    hdui = fits.open(filei)[1]
    hduz = fits.open(filez)[1]
    
    mywcs = WCS(hdug.header)
    
    image_g = hdug.data
    image_i = hdui.data
    image_z = hduz.data
    
    x_sn, y_sn = mywcs.wcs_world2pix(ra, dec, 0)
    xint = int(x_sn)
    yint = int(y_sn)
    
    width = 60
    subwidth = width*2
    subdata_g = image_g[yint-subwidth:yint+subwidth, xint-subwidth:xint+subwidth].ravel()
    subdata_i = image_i[yint-subwidth:yint+subwidth, xint-subwidth:xint+subwidth].ravel()
    subdata_z = image_z[yint-subwidth:yint+subwidth, xint-subwidth:xint+subwidth].ravel()
    
    
    xoff_arcsec = 5
    yoff_arcsec = xoff_arcsec * np.cos(dec/180*np.pi)
    ra_ref = ra-0.0032
    dec_ref = dec-0.0032* np.cos(dec/180*np.pi)
    x2c, y2c = mywcs.wcs_world2pix(ra_ref, dec_ref, 0)
    x2n, y2n = mywcs.wcs_world2pix(ra_ref, dec_ref+yoff_arcsec/3600, 0)
    x2e, y2e = mywcs.wcs_world2pix(ra_ref+xoff_arcsec/3600, dec_ref, 0)
    
    
    separation = 1/3600 * u.deg
    coo = SkyCoord(ra=ra*u.degree, dec=dec*u.degree, frame='icrs')
    
    tb = pd.read_csv('../data/data_20mrf/host/186954.csv')

    ra_host = tb["ra"].values[0]
    dec_host = tb["dec"].values[0]
    
    c_host = SkyCoord(ra=ra_host*u.degree, dec=dec_host*u.degree, frame='icrs')
    
    sep_host = coo.separation(c_host).arcsec
    print ("Source is %.4f arcsec away from the host"%sep_host)
    
    x_host, y_host = mywcs.wcs_world2pix(ra_host, dec_host, 0)
    
    dat_g = subdata_g.flatten()
    dat_i = subdata_i.flatten()
    dat_z = subdata_z.flatten()
    
    sig_g = np.median(np.abs(dat_g-np.median(dat_g)))
    sig_i = np.median(np.abs(dat_i-np.median(dat_i)))
    sig_z = np.median(np.abs(dat_z-np.median(dat_z)))
    
    multig = sig_z/sig_g-1.5
    multii = sig_z/sig_i-0.5
    min_img = np.percentile(dat_g*multig, [50])[0]
    
    image = make_lupton_rgb(image_z, image_i*multii, image_g*multig, minimum=min_img, stretch=0.8, Q=5)
```

```python
z = 0.1353
t0 = 59012.
D = cosmo.luminosity_distance([z])[0].value * 1e+6 # in pc
D_cm = D * const.pc.cgs.value
dis_mod = 5*np.log10(D/10)
dis_mod_ = dis_mod - 2.5 * np.log10(1+z)

eV = 1.60217733e-12
```

```python
def load_radio_model(ax, fit_name = 'e1_model1', color = "k"):
    print ("")
    print (fit_name)
    mcmc_dir = "../radio/mcmc_result"
    
    fname = '%s/%s_%s.pickle'%(mcmc_dir, "AT2020mrf", fit_name)
    mcmc_mod1 = pickle.load(open(fname,'rb'),encoding='latin1')[0]
    
    lgL_peak = mcmc_mod1["lgL_peak"]
    L_peak_best = 10**lgL_peak[0]
    L_peak_left = 10**(lgL_peak[0]-lgL_peak[1])
    L_peak_right = 10**(lgL_peak[0]+lgL_peak[2])
    L_peak = np.array([L_peak_best, 
                       L_peak_best-L_peak_left, 
                       L_peak_right - L_peak_best])
    print ("Lnu_peak: ", L_peak/1e+29, "e+29 erg/s")
    
    nu_peak = mcmc_mod1["nu_peak"]
    print ("nu_peak:", nu_peak)
    beta2 = mcmc_mod1["beta2"]
    print ("beta2:", beta2)
    
    lc = mcmc_mod1["lc"]
    nu = lc["nu"] * 1e+9 #  Hz
    nuLnus = lc["nuLnu"] * 1e+9 
    
    linestyle = "-."
    beta1 = mcmc_mod1["beta1"]
    print ("beta1: ", beta1)
    
    Ls = np.percentile(nuLnus, [16, 50, 84], axis = 0)
    ax.plot(nu, Ls[1] , '-', color=color, alpha=1, zorder=1,
            linestyle = linestyle)
    ax.fill_between(nu, Ls[0], Ls[2], color = color, alpha = 0.3)
    
    return nu, Ls[1]



def add_radio_data(ax, epoch = "e1"):
    if epoch == "e1":
        filename = "../data/data_20mrf/radio/radio_e1.dat"
        color = "k"
        marker = "o"
        mylabel = r"$\Delta t \approx 261$"+" days"
    elif epoch == "e2":
        filename = "../data/data_20mrf/radio/radio_e3.dat"
        color = "tab:green"
        marker = "s"
        mylabel = r"$\Delta t \approx 417$"+" days"
    tb = asci.read(filename)
    
    nu_obs = tb["nu"].data * 1e+9 # Hz
    nu = nu_obs * (1+z)
    
    fnu_obs = tb["fnu"].data * 1e-6 * 1e-23 # erg / s / cm^2 / Hz
    fnu_obs_unc = tb["fnu_unc"].data * 1e-6 * 1e-23
    
    nuLnu = (4 * np.pi * D_cm**2) * fnu_obs * nu_obs
    nuLnu_unc = (4 * np.pi * D_cm**2) * fnu_obs_unc * nu_obs
    ax.errorbar(nu, nuLnu, nuLnu_unc, fmt = marker, color = color,
                markersize = ms-2, linewidth = 0.8,
                label = mylabel)
    
    
def custom_ax(ax):
    ax.semilogx()
    ax.semilogy()
    xmin = 1e+9
    xmax = 3e+18
    ax.set_xlim(xmin, xmax)
    
    ymin = 1e+37
    ymax = 2e+43
    ax.set_ylim(ymin, ymax)  
    
    ax.set_xlabel(r"$\nu_{\rm rest}$"+" (Hz)")
    ax.set_ylabel(r"$\nu L_\nu\,({\rm erg\,s^{-1}}$"+")")
    
    ax.tick_params(which = 'major', length = 4, top=True, direction = "in", right = True)
    ax.tick_params(which = 'minor', length = 2, top=True, direction = "in", right = True)
    
    
def add_xray(ax):
    # Lnu = A nu^alpha
    
    # 36 days
    # 0.2--8 keV flux at z=0 is 4.7601e-13
    numin = 0.2 * 1e+3 * eV / const.h.cgs.value
    numax = 10 * 1e+3 * eV / const.h.cgs.value
    alpha = -(1.81-1)
    L = 3.90e-13 * 4 * np.pi * D_cm**2
    A = L * (alpha+1) / (numax**(alpha+1) - numin**(alpha+1))
    nus = np.linspace(numin, numax)
    Lnus = A * nus**(alpha)
    nuLnus = nus * Lnus
    ax.plot(nus, nuLnus, color = "r", linewidth = 2, label = r"$\Delta t \approx 36$"+" days")
    
    # 328 days
    # 1.0--10 keV flux at z=0 is 3.9466e-14 and 1.5951e-14
    multi = 1.5951e-14 /  3.9466e-14
    numin = 1.0 * 1e+3 * eV / const.h.cgs.value
    numax = 10 * 1e+3 * eV / const.h.cgs.value
    alpha = -(0.99-1)
    L = 3.9466e-14 * 4 * np.pi * D_cm**2
    A = L * (alpha+1) / (numax**(alpha+1) - numin**(alpha+1))
    nus = np.linspace(numin, numax)
    Lnus = A * nus**(alpha)
    nuLnus = nus * Lnus
    ax.fill_between(nus, nuLnus, nuLnus*multi, alpha = 0.3,
                    color = "b", label = r"$\Delta t \approx 328$"+" days")
    
    
def add_opt(ax):
    nu_pool = {'g.ztf': 634121907623677.2,
               'r.ztf': 471994480402802.2,
               'i.ztf': 380021535091050.1,
               'o.atlas': 450003936111481.4,
               'c.atlas': 575819723942441.0}
    
    name = "ZTF20abfhyil"
    fname = '../data/data_20mrf/phot/%s_all_lc.dat'%(name)
    lc_data = asci.read(fname)

    # convert to rest-frame days
    tpeak_mjd = 59016.00
    dt_all = np.array((lc_data['mjd'] - tpeak_mjd) /(1+z) )
    nu_all_obs = np.array([nu_pool[x['filter']] for x in lc_data])
    nu_all = nu_all_obs * (1+z) # rest-frame wavelength used for fit

    # convert to luminosity: nuLnu 
    # not that this quantity is the same in either rest-frame or observer's frame
    lum_all = np.array(flux2lum(lc_data["uJy.0"], nu_all_obs, z))
    e_lum_all = np.array(flux2lum(lc_data["duJy.0"], nu_all_obs, z))
    
    ix = (dt_all > 34.6)&(dt_all < 37.5)&(e_lum_all < lum_all)
    
    for myfilt in ["r.ztf", "g.ztf",  "o.atlas", "c.atlas"]:
        ixband = ix & (nu_all_obs == nu_pool[myfilt])
        if np.sum(ixband)>0:
            L = lum_all[ixband]
            eL = e_lum_all[ixband]
            nu = nu_pool[myfilt]
            
            if np.sum(ixband)==1:
                ll = L[0]
                ell = eL[0]
            else:
                print (myfilt)
            
            ax.errorbar(nu, ll, ell, color = "red", fmt = "o")
            
            print (myfilt, nu, ll, ell)
        
    
def plot_multiband():
    plt.figure(figsize = (4.8, 3.2))
    ax = plt.subplot(111)
    add_opt(ax)
    
    # radio emission at 261 days
    add_radio_data(ax, epoch = "e1")
    add_radio_data(ax, epoch = "e2")
    
    custom_ax(ax)
    
    mynu, Ls1 = load_radio_model(ax, fit_name = 'e1_model3', color = "gray")
    mynu, Ls3 = load_radio_model(ax, fit_name = 'e3_model3', color = "tab:green")
    
    t1 = 328 - 261
    t2 = 417 - 328
    wtg1_ = 1/t1
    wtg2_ = 1/t2
    wtg1 = wtg1_ / (wtg1_ + wtg2_ )
    wtg2 = wtg2_ / (wtg1_ + wtg2_ )
    
    Ls2 = Ls1*wtg1 + Ls3 * wtg2
    ax.plot(mynu, Ls2, color = "b", linestyle = "--")
    
    # Berger+2012, Zauderer+2013
    nu = np.logspace(10, 19, 2000)
    nu_peak = 10**9.62
    s = 1.
    beta1 = 2
    beta2 = -1*(2.8-1)/2
    Lnu1 = 2e+29 * ((nu/nu_peak)**(-s * beta1) + (nu/nu_peak)**(-s * beta2))**(-1/s)
    nuLnu1 = nu*Lnu1
    
    B = 0.2 # Gauss
    nu_g =  const.e.gauss.value * B / (2 * np.pi * const.m_e.cgs.value * const.c.cgs.value)
    temp = (6 * np.pi * const.m_e.cgs.value * const.c.cgs.value)
    gamma_c = temp / (const.sigma_T.cgs.value * B**2 * (317 * 24 * 3600))
    nu_c = gamma_c**2 * nu_g
    print ("nu_c = %.2f e+11 Hz"%(nu_c/1e+11))
    nu_c = 3e+11
    
    nu_peak = nu_c
    s = 5
    beta1 = -1*(2.8-1)/2
    beta2 = -2.8/2
    Lnu2 = 10**27.64 * ((nu/nu_peak)**(-s * beta1) + (nu/nu_peak)**(-s * beta2))**(-1/s)
    nuLnu2 = nu*Lnu2
    nuLnu = nuLnu1
    ix = nu>1e+11
    nuLnu[ix] = nuLnu2[ix]
    
    ix = nu>mynu[-1]
    ax.plot(nu[ix], nuLnu[ix], color ="b", linestyle = "--")
    
    
    add_xray(ax)
    
    ax.legend()
    plt.tight_layout(rect =(-0.03, -0.05, 1.02, 1.02))
```

```python
plot_multiband()
```

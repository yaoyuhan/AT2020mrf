```python
import pickle
import numpy as np
import astropy.io.ascii as asci
import astropy.constants as const

from scipy.interpolate import interp1d

import matplotlib
import matplotlib.pyplot as plt
fs= 10
matplotlib.rcParams['font.size']=fs
ms = 6
matplotlib.rcParams['lines.markersize']=ms

from astropy.cosmology import FlatLambdaCDM
cosmo = FlatLambdaCDM(H0=70., Om0=0.3)

z = 0.1353
t0 = 59012.
D = cosmo.luminosity_distance([z])[0].value * 1e+6 # in pc
D_cm = D * const.pc.cgs.value

DA = cosmo.angular_diameter_distance(z).value * 1e+6 # in pc
```

```python
def add_data_e1(ax, color = "k"):
    filename = "../data/data_20mrf/radio/radio_e1.dat"
    tb = asci.read(filename)
    
    nu_obs = tb["nu"].data
    nu = nu_obs * (1+z)
    
    fnu_obs = tb["fnu"].data * 1e-6 * 1e-23
    fnu_obs_unc = tb["fnu_unc"].data * 1e-6 * 1e-23
    
    nuLnu = (4 * np.pi * D_cm**2) * fnu_obs * nu_obs
    nuLnu_unc = (4 * np.pi * D_cm**2) * fnu_obs_unc * nu_obs
    marker = "o"
    ax.errorbar(nu, nuLnu/nu, nuLnu_unc/nu, fmt = marker, color = color,
                 markersize = ms-2, linewidth = 0.8,label = r"$\Delta t \approx 261$"+" days")
    ax.tick_params(which = 'major', length = 4, top=True, direction = "in", right = True)
    ax.tick_params(which = 'minor', length = 2, top=True, direction = "in", right = True)


def add_data_e2(ax, color = "tab:blue"):
    filename = "../data/data_20mrf/radio/radio_e2.dat"
    tb = asci.read(filename)
    
    nu_obs = tb["nu"].data
    nu = nu_obs * (1+z)
    
    fnu_obs = tb["fnu"].data * 1e-6 * 1e-23
    fnu_obs_unc = tb["fnu_unc"].data * 1e-6 * 1e-23
    
    nuLnu = (4 * np.pi * D_cm**2) * fnu_obs * nu_obs
    nuLnu_unc = (4 * np.pi * D_cm**2) * fnu_obs_unc * nu_obs
    
    ixdet = nuLnu_unc!=0
    marker = "d"
    ax.errorbar(nu[ixdet], nuLnu[ixdet]/nu[ixdet], nuLnu_unc[ixdet]/nu[ixdet], 
                fmt = marker, color = color, linewidth = 0.8,
                markersize = ms-2, label = r"$\Delta t \approx 305$"+" days")
    ax.plot(nu[~ixdet], nuLnu[~ixdet]/nu[~ixdet], color = color, marker = "v",
            markersize = ms-1)
    
    
def add_data_e3(ax, color = "tab:green"):
    filename = "../data/data_20mrf/radio/radio_e3.dat"
    tb = asci.read(filename)
    
    nu_obs = tb["nu"].data
    nu = nu_obs * (1+z)
    
    fnu_obs = tb["fnu"].data * 1e-6 * 1e-23
    fnu_obs_unc = tb["fnu_unc"].data * 1e-6 * 1e-23
    
    nuLnu = (4 * np.pi * D_cm**2) * fnu_obs * nu_obs
    nuLnu_unc = (4 * np.pi * D_cm**2) * fnu_obs_unc * nu_obs
    
    ixdet = nuLnu_unc!=0
    marker = "s"
    ax.errorbar(nu[ixdet], nuLnu[ixdet]/nu[ixdet], nuLnu_unc[ixdet]/nu[ixdet], 
                fmt = marker, color = color, linewidth = 0.8,
                markersize = ms-2, label = r"$\Delta t \approx 417$"+" days")
```

```python
def load_models(ax, fit_name = 'e1_model1', color = "k"):
    print ("")
    print (fit_name)
    mcmc_dir = "./mcmc_result"
    
    fname = '%s/%s_%s.pickle'%(mcmc_dir, "AT2020mrf", fit_name)
    mcmc_mod1 = pickle.load(open(fname,'rb'),encoding='latin1')[0]
    
    lgL_peak = mcmc_mod1["lgL_peak"]
    L_peak_best = 10**lgL_peak[0]
    L_peak_left = 10**(lgL_peak[0]-lgL_peak[1])
    L_peak_right = 10**(lgL_peak[0]+lgL_peak[2])
    L_peak = np.array([L_peak_best, 
                       L_peak_best-L_peak_left, 
                       L_peak_right - L_peak_best])
    print ("Lnu_peak: ", L_peak/1e+29, "e+29 erg/s")
    
    nu_peak = mcmc_mod1["nu_peak"]
    print ("nu_peak:", nu_peak)
    beta2 = mcmc_mod1["beta2"]
    print ("beta2:", beta2)
    
    beta2str = r"$\beta_2 = %.1f_{-%.1f}^{+%.1f}$"%(beta2[0], beta2[1], beta2[2])
    
    lc = mcmc_mod1["lc"]
    nu = lc["nu"]
    nuLnus = lc["nuLnu"]
    
    if fit_name == 'e1_model2':
        linestyle = "--"
        beta1str = r"$\beta_1 = 2.5$"+" (fixed), "
    else:
        linestyle = "-."
        beta1 = mcmc_mod1["beta1"]
        print ("beta1: ", beta1)
        beta1str = r"$\beta_1 = %.1f_{-%.1f}^{+%.1f}$, "%(beta1[0], beta1[1], beta1[2])
    if "e3" in fit_name:
        linestyle = ":"
    mylabel = beta1str+beta2str
    
    lgs = mcmc_mod1["lgs"]
    s = 10**lgs
    print ("s:", s)
    
    Ls = np.percentile(nuLnus, [16, 50, 84], axis = 0)
    ax.plot(nu, Ls[1] / nu, color=color, zorder=1,
            linestyle = linestyle, label = mylabel, alpha = 0.7)
    ax.fill_between(nu, Ls[0]/ nu, Ls[2]/ nu, color = color, alpha = 0.2)
```

```python
def custom_ax(ax, xmin = 1, ymin = 1e+28):
    ax.semilogx()
    ax.semilogy()
    xmax = 35
    ax.set_xlim(xmin, xmax)
    ymax = 2e+29
    ax.set_ylim(ymin, ymax)  
    ax.set_xlabel(r"$\nu_{\rm rest}$"+" (GHz)")
    ax.set_ylabel(r"$L_\nu\,({\rm erg\,s^{-1}\,Hz^{-1}}$"+")")
```

```python
def add_18cow_n21(ax):
    z = 0.014
    D = cosmo.luminosity_distance([z])[0].value * 1e+6 # in pc
    D_cm = D * const.pc.cgs.value
    
    filename1 = "../data/data_cow/radio/Nayana21_tab1.dat"
    tb1 = asci.read(filename1, delimiter = "\t")
    tb1 = tb1[2:]
    z = 0.014
    D = cosmo.luminosity_distance([z])[0].value * 1e+6 # in pc
    D_cm = D * const.pc.cgs.value
    
    nu_obs = tb1["Frequency"].data
    t_obs = tb1["Age"].data
    
    temp = tb1["Flux Density"].data
    fnu_obs = np.array([int(x.split(" +or-")[0]) for x in temp]) * 1e-6 * 1e-23
    fnu_obs_unc = np.array([int(x.split("+or-")[1]) for x in temp]) * 1e-6 * 1e-23
    Lnu_obs = fnu_obs * 4 * np.pi * D_cm**2
    Lnu_obs_unc = fnu_obs_unc * 4 * np.pi * D_cm**2
    
    for mynu in [0.4, 0.75, 1.25]:
        ind = nu_obs == mynu
        myfunc = interp1d(t_obs[ind], Lnu_obs[ind])
        emyfunc = interp1d(t_obs[ind], Lnu_obs_unc[ind]*2)
        
        t_haha = [261, 305, 417]
        color_haha = ["k", "tab:blue", "tab:green"]
        marker_haha = ["o", "d", "s"]
        for i in range(3):
            tt = t_haha[i]
            color = color_haha[i]
            marker = marker_haha[i]
            LL = myfunc(tt)
            eLL = emyfunc(tt)
            eLL = max(0.15*LL, eLL)
            #print (LL)
            ax.errorbar(mynu*(1+z), LL, eLL, 
                        markeredgecolor = color, ecolor=color,
                        markerfacecolor = "none",
                        markersize = ms, marker = marker)
    
```

```python
def load_model_inhomo(ax):
    print ("")
    fit_name = 'model_ssa'
    mcmc_dir = "./inhomo_result"
    print (fit_name)
    
    fname = '%s/%s_%s.pickle'%(mcmc_dir, "AT2020mrf", fit_name)
    mcmc_mod1 = pickle.load(open(fname,'rb'),encoding='latin1')[0]
    
    for ycol in ["lgK1", "lgK2", "lgsigma0"]:
        lgL_peak = mcmc_mod1[ycol]
        L_peak_best = 10**lgL_peak[0]
        L_peak_left = 10**(lgL_peak[0]-lgL_peak[1])
        L_peak_right = 10**(lgL_peak[0]+lgL_peak[2])
        L_peak = np.array([L_peak_best, 
                           L_peak_best-L_peak_left, 
                           L_peak_right - L_peak_best])
        print (ycol[2:], ":", L_peak)
    
    for ycol in ["beta", "a", "b", "p"]:
        par = mcmc_mod1[ycol]
        print (ycol, ":",  par)
    
    lc = mcmc_mod1["lc"]
    nu = lc["nu"]
    t = lc["t"]
    nuLnus = lc["nuLnu"]
    
    Ls = np.percentile(nuLnus, [16, 50, 84], axis = 0)
    
    nxl = int(len(t)/3)
    
    colors = ["k", "tab:blue", "tab:green"]
    linestyle = "--"
    for j in range(3):
        color = colors[j]
        ind = np.arange(nxl*j, nxl*(j+1))
        ax.plot(nu[ind], Ls[1][ind] / nu[ind], color=color, zorder=1,
                linestyle = linestyle, alpha = 0.7)
        ax.fill_between(nu[ind], Ls[0][ind]/ nu[ind], Ls[2][ind]/ nu[ind], color = color, alpha = 0.2)
```
